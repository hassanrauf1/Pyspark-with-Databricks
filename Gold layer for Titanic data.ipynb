```python
df_silver = spark.table("workspace.default.titanic_silver")
```

```python
from pyspark.sql.functions import count, sum

df_gold = df_silver.groupBy(
    "Sex",
    "Age_Group",
    "Pclass"
).agg(
    count("*").alias("total_passengers"),
    sum("Survived").alias("total_survived")
)
```

```python
from pyspark.sql.functions import col

df_gold = df_gold.withColumn(
    "survival_rate",
    col("total_survived") / col("total_passengers")
)
```

```python
from pyspark.sql.functions import count, sum, col, round

gender_survival = df_silver.groupBy("Sex").agg(
    count("*").alias("total_passengers"),
    sum("Survived").alias("total_survived")
).withColumn(
    "survival_percentage",
    round((col("total_survived") / col("total_passengers")) * 100, 2)
)

display(gender_survival)
```

```python
age_survival = df_silver.groupBy("Age_Group").agg(
    count("*").alias("total_passengers"),
    sum("Survived").alias("total_survived")
).withColumn(
    "survival_percentage",
    round((col("total_survived") / col("total_passengers")) * 100, 2)
)

display(age_survival)
```

```python
display(gender_survival.orderBy(col("survival_percentage").desc()))
display(age_survival.orderBy(col("survival_percentage").desc()))
```

```python
gender_survival.write.mode("overwrite").format("delta") \
    .saveAsTable("workspace.default.titanic_gold_gender")

age_survival.write.mode("overwrite").format("delta") \
    .saveAsTable("workspace.default.titanic_gold_age")
```

```python
gold_dashboard = df_silver.groupBy(
    "Sex",
    "Age_Group",
    "Pclass"
).agg(
    count("*").alias("total_passengers"),
    sum("Survived").alias("total_survived")
).withColumn(
    "survival_rate",
    round((col("total_survived") / col("total_passengers")) * 100, 2)
)

gender_survival.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.default.titanic_gold_gender")

age_survival.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.default.titanic_gold_age")

gold_dashboard.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.default.titanic_gold_dashboard")
```
